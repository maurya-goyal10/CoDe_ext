```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from pathlib import Path
```

```python
db = np.load('./embeddings/csd_vit_large_custom_normal/1/database/embeddings_0.pkl', allow_pickle=True)
query = np.load('./embeddings_code40_b5_r6_s1/csd_vit_large_custom_normal/1/database/embeddings_0.pkl', allow_pickle=True)
```

```python
db['filenames']
```

```python
query['filenames']
```

```python
style_folder = Path('./style_folder/styles')
img_folder = Path('../BoN/outputs/cherry_picking/c_code40_b5_r6_styletransfer/style_1/images/A colorful photo of a eiffel tower')

fig, axes = plt.subplots(1, len(query['filenames'])+1, figsize=(20,5))

for i, ax in enumerate(axes.flatten()):
    if i == 0:
        ax.imshow(Image.open(Path(style_folder).joinpath(db['filenames'][2] + '.jpg')))
        ax.set_title('Ref.')

    else:
        ax.imshow(Image.open(Path(img_folder).joinpath(query['filenames'][i-1] + '.png')))
        score = query['embeddings'][i-1] @ db['embeddings'][2].T
        ax.set_title(f'{query["filenames"][i-1]}: {score}')

    ax.axis('off')

```
